```python
fig1=plt.figure(figsize=(5,4))
ax1=fig1.add_axes([1,1,1,1])

plt.title("company profit for month")
plt.xlabel("month number")
plt.ylabel("total profit")

ax1.set_ylim(100000,500000)
ax1.set_xlim(0.5,12.5)

x=df["month_number"]
y0=df["total_profit"]

ax1.plot(x,y0)

plt.show()
```

```python
import matplotlib.colors as mcolors

y1=df["facecream"]
y2=df["facewash"]
y3=df["toothpaste"]
y4=df["bathingsoap"]
y5=df["shampoo"]
y6=df["moisturizer"]

fig2=plt.figure(figsize=(5,4))
ax2=fig2.add_axes([1,1,1,1])


plt.title("sales data")
plt.xlabel("Month number")
plt.ylabel("sales units in number")

ax2.set_xlim(0.5,12.5)
ax2.set_ylim(800,18000)

ax2.plot(x,y1,'-o',color='blue',linewidth=3)
ax2.plot(x,y2,'-o',color='orange',linewidth=3)
ax2.plot(x,y3,'-o',color='green',linewidth=3)
ax2.plot(x,y4,'-o',color='red',linewidth=3)
ax2.plot(x,y5,'-o',color='purple',linewidth=3)
ax2.plot(x,y6,'-o',color='brown',linewidth=3)

plt.show()
```

```python
fig3=plt.figure(figsize=(5,4))
ax3=fig3.add_axes([1,1,1,1])
plt.title("Tooth paste Sales data")
plt.xlabel("Month number")
plt.ylabel("Number of units Sold")

ax3.plot(x,y3,'o')
plt.grid(linestyle='--')
plt.legend(['Tooth paste Sales data'],loc=2)
plt.show()
```

```python
fig4=plt.figure(figsize=(5,4))
ax4=fig4.add_axes([1,1,1,1])

plt.title("Facewash and facecream sales data")
plt.xlabel("Month Number")
plt.ylabel("Sales units in number")


width = 0.35

ax4.bar(x-width/2,y1,width,align='center',label='Face Cream sales data')#you can add color='blue' parameter
ax4.bar(x+width/2,y2,width,color='orange',align='center',label='Face Wash sales data')
#matplotlib.pyplot.bar(x, height, width=0.8, bottom=None, *, align='center', data=None, **kwargs)etc:xerr,yerr[source]

plt.grid(linestyle='--')
plt.legend(loc=2)
#plt.legend(['Face Wash sales data','Face Wash sales data'], loc=2)
plt.show()
```

```python
fig5=plt.figure(figsize=(5,4))
ax5=fig5.add_axes([1,1,1,1])
plt.title("Sales data")

pie=[]

products=["FaceCream","FaceWash","Toothpaste","Bathingsoap","Shampoo","Moisturizer"]

pie.append(y1.sum())
pie.append(y2.sum())
pie.append(y3.sum())
pie.append(y4.sum())
pie.append(y5.sum())
pie.append(y6.sum())

ax5.pie(pie,labels=products,autopct='%1.2f%%')
plt.legend(["FaceCream","FaceWash","Toothpaste","Bathingsoap","Shampoo","Moisturizer"],loc=4)

plt.show()
```

```python


fig6=plt.figure(figsize=(5,4))
ax6=fig6.add_axes([1,1,1,1])

plt.title("All product sales data using stack plot")
plt.xlabel("Month number")
plt.ylabel("Sales units in Number")

#ax1.fill_between(x, y_stack[1,:], y_stack[2,:], facecolor="#6E5160")

ax6.stackplot(x,y1,y2,y3,y4,y5,y6)

ax6.legend(['FaceCream','FaceWash','Toothpaste','Bathingsoap','Shampoo','Moisturizer'],loc=2)

plt.show()
```
